La bibliote Matplotlib, sirve para realizar gráficas, como seaborn.
Pueden ser en 2D, 3D, y tiene una gran variedad de grafos.

```python
import matplotlib.pyplot as plt
import numpy as np
import math
#Graficar un arreglo de puntos
xpoints = [1,2,3,4]
ypoints = [5,6,7,8]
plt.plot(xpoints,ypoints)
plt.show()
```

```python
xpoints = np.linspace(0,10,100) #Esta es la función en x, se agarra un intervalo del 0 al 10 y lo divide en 100
ypoints = np.sin(xpoints)

plt.plot(xpoints,ypoints, color='blue')
plt.title("Gráfica de la función seno") #Título del gráfico
plt.xlabel("Eje x") #El nombre (valores) del eje "x"
plt.ylabel("Eje y") #El nomnre (valores) del eje "y" y la función evaluada en x
plt.grid(True) #Para mostrar la malla
plt.show()
```

```python
xpoints = np.linspace(0,10,100) #Esta es la función en x, se agarra un intervalo del 0 al 10 y lo divide en 100
ypoints = np.sin(xpoints)

plt.plot(xpoints,ypoints, color='red', linewidth=1, linestyle='--') #linewidth es el ancho o grosor de la linea, y linestyle es el estilo de la línea (discontinua, dispersión, etc.)
plt.title("Gráfica de la función seno") #Título del gráfico
plt.xlabel("Eje x") #El nombre (valores) del eje "x"
plt.ylabel("Eje y") #El nomnre (valores) del eje "y" y la función evaluada en x
plt.grid(True) #Para mostrar la malla
plt.show()
```

```python
xpoints = np.linspace(0,10,100) #Esta es la función en x, se agarra un intervalo del 0 al 10 y lo divide en 100
ypoints = np.cos(xpoints)

plt.scatter(xpoints,ypoints, color='magenta', marker='x', label='datos') #marker, significa un símbolo para los datos discretos
plt.lagend() #pone leyenda (información de los datos) dentro del grafo

plt.title("Gráfica de la función coseno")
plt.xlabel("Eje x") #El nombre (valores) del eje "x"
plt.ylabel("Eje y") #El nomnre (valores) del eje "y" y la función evaluada en x
plt.grid(True) #Para mostrar la malla
plt.show()
```

```python
#Ejercicio de poner dos gráficas juntas
x=np.linspace(-10,10,100) #siempre se comienza definiendo el np.linspace para el intervalo de x
y1=np.sin(x)
y2=np.cos(x)

plt.figure(figsize=(10,5))  #Es para el tamaño de la gráfica (figura), dice que el ancho es dos veces más grande que el alto
#El primer valor es el ancho de la figura, el segundo valor es el alto de la figura
plt.plot(x, y1, label='sin(x)', color='green')
plt.plot(x, y2, label='cos(x)', color='red')
plt.legend()
plt.title("Gráfica de la función seno y coseno")
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.grid(True)
plt.show
#Aquí las graficas estan encimadas
```

```python
#Para tener un mosaico de gráficas, se utiliza plt.subplot(a,b,p)
# a es las filas, b es las columnas, p es el número posición de la gráfica

xn=np.linspace(-10,10,100)

plt.figure(figsize=(10,9))

#subgráfica 1
plt.subplot(2,2,1)
plt.plot(xn, np.cos(xn), color='brown')
plt.title('Gráfica coseno')
plt.grid()

#subgráfica 2
plt.subplot(2,2,2)
plt.plot(xn, np.sin(xn), color='blue')
plt.title('Gráfica seno')
plt.grid()

#subgráfica 3
plt.subplot(2,2,3)
plt.plot(xn, np.tan(xn), color='magenta')
plt.title('Gráfica tangente')
plt.grid()

#subgráfica 4
plt.subplot(2,2,4)
plt.plot(xn, xn**2, color='red')
plt.title('Gráfica x cuadrada')
plt.grid()

plt.show()
```

Gráfica de barras e Histograma

```python
etiquetas=['A', 'B', 'C', 'D']
valor=[20, 50, 95, 18]

plt.bar(etiquetas, valor, color='violet') #Es para datos discretos "bar"
plt.title("Gráfica de barras")
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.grid(True)
plt.show()
```

```python
datos=np.random.randint(1,100, size=100)

plt.hist(datos, bins=10, color='magenta', edgecolor='black')  #bist es el número de barras que se va a ir acumulando
plt.title("Histograma")
plt.xlabel("Eje x")
plt.ylabel("Eje y")
#plt.grid(True)
plt.show()
```

```python
#Ahora para que nos muestre los valores arriba de las barras
datos=np.random.randint(1,100, size=100)

cuentas, bins, patches=plt.hist(datos, bins=15, color='blue', edgecolor='black', rwidth=0.9) #rwidth es para espaciar las barras
#cuentas es el valor de cada barra del histograma (bins), se hace un ciclo for para conocer sus valores
#A continuación vamos a agregar etiquetas (valores)

for count, x in zip(cuentas, bins[:-1]): #Por cada bints le voy a sumar elvalor de x más su promedio
  plt.text(x+(bins[1] - bins[0])/2, count, str(int(count)), #"plt.text(x+(bins[:-1] - bins[0])/2" es para graficar los valores y define que es para que este medio de la barra.
  ha='center', va='bottom', fontsize=10, fontweight='bold') #Los font son para el tamaño y tipo de letra.

plt.title("Histograma")
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.grid(axis='y', linestyle='--', alpha=0.6) #solo se pone la malla para el eje y, con trasparencia de 0.6
plt.show()
```

```python
#Gráfica de pastel
frutas=['Manzanas', 'Plátanos', 'Fresas', 'Uvas']
cantidad=[40,35, 15, 10]

plt.pie(cantidad,labels=frutas, autopct='%1.1f%%', startangle=90, shadow=True)
plt.title("Porcentaje de frutas")
plt.show()
```
