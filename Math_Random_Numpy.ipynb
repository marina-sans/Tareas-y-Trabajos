Random

```python
#Primero se importa, y sirve para obtener un numero aleatorio y realizar análisis estadístico, y reaizar encriptaciones.
import random
random.seed(1)
print(random.random())
```

```python
import random
mylist=["manzana", "pera","chayote"]
random.choice(mylist)
```

```python
#Pueden salir los mismos parámetros repetidos
import random
mylist=["manzana", "pera","chayote", "berengena", "papaya"]
random.choices(mylist, k=3)
```

```python
# "sample" no permite repeticiones
random.sample(mylist, k=3)
```
