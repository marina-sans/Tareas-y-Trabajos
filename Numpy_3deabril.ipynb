Arreglos bidimencionales

```python
B = np.array([[1,2,3], [4,5,6]]) #Aqui los elementos son los bloquecitos:
print(B[0])
```

```python
print(B[1][1]) #Agarra, del bloque 1 al elemento 1
```

```python
print(B[:])
```

```python
print(B[:,1]) #De todos los arreglos se agarra el elemento 1
```

```python
print(B[:][1]) #Me tomo los dos arreglos y agarro el número 1
#Siempre que se seleccionen rebanadas se usa la coma
```

```python
#Seleccionas 1,2 y 4,5
print(B[:,0:2])
```

```python
print(B[:,::2])
```

```python
print(B[0,0:2]) #Lo que va después de los puntitos es para delimitar, y luego el rango de donde se va a agarrar
```

Factorial

```python
#Factorial iterativo, hecho en clase:
def factorial(n):
  if type(n)==int:
    fact=1 #Es un contador
    for i in range(n):
      fact=fact*(i+1) #Le sumamos 1 porque empieza en 0, y así hace que termine en el número que pedimos

    return fact
  else:
    print("Inválido")
factorial("J")
#Para comentar el código podemos usas "DeepSeek"
```
